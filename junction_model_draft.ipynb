```python
import numpy as np
from scipy.constants import e, pi, m_e, h
from scipy.sparse.linalg import lsmr
from scipy.sparse import csr_matrix

class MemristorNetwork:
    def __init__(self, adjacency_matrix, lambda_initial, dt=0.001):
        self.adj = adjacency_matrix
        self.lambda_vals = lambda_initial
        self.dt = dt
        self.m_star = 0.99 * m_e
        self.phi = 0.82 #* e
        self.lambda_crit = 0.01
        self.s_max = 5e-9
        self.A = (0.41e-9)**2
        self.R_on = 13.0e3
        self.R_off = 13.0e6
        self.V_set = 1.0e-2
        self.V_reset = 1.0e-3
        self.b = 10.0
        self.lambda_max = 0.015
        self.G = self.calculate_conductance_matrix()
        self.branch_currents = np.zeros(self.adj.shape)
        # self.W = np.multiply(self.adj, self.G)

    def constrain_lambda(self, lambda_val, lambda_min, lambda_max):
        return np.clip(lambda_val, lambda_min, lambda_max)
        
    def calculate_conductance_matrix(self):
        G = np.zeros(self.adj.shape)
        for i in range(len(self.lambda_vals)):
            for j in range(len(self.lambda_vals)):
                if self.adj[i, j] != 0:
                    G[i, j] = self.G_lambda(self.lambda_vals[i, j])
        return G
    
    def G_lambda(self, lambda_param):
        s = max((self.lambda_crit - np.abs(lambda_param)) * self.s_max / self.lambda_crit, 0)
        if s > 0:
            G_t = 3 * (2 * self.m_star)**0.5 * e**2.5 * (self.phi)**0.5 / (self.A * 2 * h**2 * s**2) * np.exp(-4 * pi * (2 * self.m_star * e)**0.5 * s * (self.phi)**0.5 / h)
        else:
            G_t = 0
        
        R_t = 1 / G_t if G_t > 0 else self.R_off
        xi = self.R_on**2 / (self.R_off - self.R_on)
        G_total = 1 / (R_t + self.R_on + xi) + 1 / self.R_off
        return G_total
    
    def d_lambda_d_t(self, v_t, lambda_t):
        if np.abs(v_t) > self.V_set:
            return (np.abs(v_t) - self.V_set) * np.sign(v_t)
        if self.V_reset < np.abs(v_t) and np.abs(v_t) < self.V_set:
            return 0
        if np.abs(v_t) < self.V_reset:
            return self.b * (np.abs(v_t) - self.V_reset) * np.sign(lambda_t)
    
    def calculate_laplacian(self):
        D = np.diag(np.sum(self.W, axis=1))
        L = D - self.W
        return L, D
    
    def solve_voltage_distribution(self, L_dagger, I):
        # print(I.shape)
        V = lsmr(csr_matrix(L_dagger), I)[0]
        return V[:-1]
    
    def calculate_branch_currents(self, V_distribution):
        """
        Calculate the currents through each branch based on the voltage distribution.
        """
        
        for i in range(len(V_distribution)):
            for j in range(i + 1, len(V_distribution)):
                if self.adj[i, j] != 0:  # Only calculate for connected nodes
                    # Current = Conductance * Voltage Difference
                    b_curr = self.G[i, j] * (V_distribution[i] - V_distribution[j])
                    self.branch_currents[i, j] = b_curr
                    # print(self.G[i, j] * (V_distribution[i] - V_distribution[j]))
                    self.branch_currents[j, i] = -b_curr  # Current direction is opposite for node j to i  
        
    
    def simulate_step(self, C, I):
        # Recalculate W as the element-wise multiplication of adjacency matrix and G
        self.W = np.multiply(self.adj, self.G)
        
        # Calculate the Laplacian and expanded Laplacian
        L, D = self.calculate_laplacian()
        
        L_dagger = self.expanded_laplacian(L, C)
        
        # Solve for voltage distribution
        V_distribution = self.solve_voltage_distribution(L_dagger, I)
        
        
        # Update lambda values based on voltage
        for i in range(len(self.lambda_vals)):
            for j in range(len(self.lambda_vals)):
                if self.adj[i, j] != 0:
                    lambda_update = self.d_lambda_d_t(V_distribution[i] - V_distribution[j], self.lambda_vals[i, j])
                    self.lambda_vals[i, j] += lambda_update * self.dt
                    self.lambda_vals[i, j] = self.constrain_lambda(self.lambda_vals[i, j], -self.lambda_max, self.lambda_max)
        
        self.G = self.calculate_conductance_matrix()
       
        # Calculate branch currents based on the new voltage distribution
        self.calculate_branch_currents(V_distribution)
        
        
        return V_distribution, self.branch_currents
    
    def expanded_laplacian(self, L, C):
        C = np.expand_dims(C, axis=-1)
        L_dagger = np.block([
            [L, C],
            [C.T, np.zeros((C.T.shape[0], C.T.shape[0]))]
        ])
        return L_dagger
```

```python
# Regenerating a larger, more complex graph for demonstration
G_demo = nx.erdos_renyi_graph(n=10, p=0.5)  # Generate a random graph
pos_demo = nx.spring_layout(G_demo)  # Position nodes using the spring layout
```

```python

# Initialization as before
# adjacency_matrix = np.array([
#     [0, 1, 0, 0],
#     [1, 0, 1, 0],
#     [0, 1, 0, 1],
#     [0, 0, 1, 0]
# ])


# Extract the adjacency matrix as a SciPy sparse matrix
adj_matrix_sparse = nx.adjacency_matrix(G_demo)

# Convert to a regular NumPy array for easier use
adjacency_matrix = adj_matrix_sparse.toarray()


lambda_initial = np.zeros_like(adjacency_matrix, dtype=float)
for i in range(len(adjacency_matrix)):
    for j in range(len(adjacency_matrix)):
        if adjacency_matrix[i, j] != 1:
            lambda_initial[i, j] = 0.0001  # Initial lambda values for existing connections
            # lambda_initial[i, j] = lambda_values_over_time[-1, i, j]

            
dt = 0.001  # Time step
# C = np.array([1, 0, 0, 1])  # Connection to external electrodes

terminal_1 = 1
terminal_2 = 3
C = np.array([0]*adjacency_matrix.shape[0])
C[terminal_1-1] = 1
C[terminal_2-1] = 1
base_I = np.array([0]*(adjacency_matrix.shape[0]+1))
base_I[terminal_1-1] = -1
base_I[terminal_2-1] = 1
# base_I = np.array([1, 0, 0, -1, 0])  # Base injected current without gain

# Define square pulse parameters
current_gain_base = 0.00  # Base current gain
pulse_gain = 0.001  # Gain during pulse

# Pulse 1 parameters
pulse1_start = 0
pulse1_duration = 2 

# Pulse 2 parameters
pulse2_start = 5  # Start after the first pulse ends
pulse2_duration = 2

num_steps = 10  # Total number of simulation steps to cover both pulses

# Initialize the MemristorNetwork
network = MemristorNetwork(adjacency_matrix, lambda_initial, dt)

# Initialize arrays or lists to store simulation data
lambda_values_over_time = np.zeros((num_steps, len(adjacency_matrix), len(adjacency_matrix)))
voltage_distributions_over_time = np.zeros((num_steps, len(adjacency_matrix)))

# I = current_gain_base * base_I
for step in range(num_steps):
    # Determine current gain based on the step, considering both pulses
    if pulse1_start <= step < pulse1_start + pulse1_duration or \
       pulse2_start <= step < pulse2_start + pulse2_duration:
        current_gain = pulse_gain
    else:
        current_gain = current_gain_base

    I = current_gain * base_I  # Update current based on the current gain
    I[terminal_1] = current_gain * 1
    I[terminal_2] = current_gain * -1
    
    # Simulate a step
    V_distribution, I_dist = network.simulate_step(C, I)
    # print(V_distribution)
    # print(I_dist)
    
    
    
    # Store the voltage distribution for analysis
    voltage_distributions_over_time[step] = V_distribution
    
    # Store the lambda values for analysis
    lambda_values_over_time[step] = network.lambda_vals.copy()
```
